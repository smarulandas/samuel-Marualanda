**x(t)= 200cos(100t) − 50sin(200t+5)**

Ejercicio:

- Repetir según ejemplo

- Repetir para entrada de -3.3 a 3.3 con conversor de 4 bits

```python
import numpy as np
#Se analizara la función en dos partes, primero 200cos(100t), donde:

A_1 = 200 #Amplitud de la función
W_1 = 100 #Frecuencia angular

#Para hallar T = (2*pi)/W_1 -> [T = Periodo]
T_1 = (2*np.pi)/(W_1)

print("A_1:",A_1," W_1:",W_1, " T_1:",T_1)

#Para hallar F = 1/T_1 -> [F = Frecuencia]
F_1 = 1/T_1

print("F_1:",F_1)

#Ahora, Se analizara la función en la segunda parte, 50sin(200t+5), donde:
A_2 = 50
W_2 = 200
U = 5 #Fase
#Para hallar T = (2*pi)/W_2
T_2 = (2*np.pi)/(W_2)

print("A_2:",A_2," W_2:",W_2, " T_2:",T_2)

#Para hallar F = 1/T_1
F_2 = 1/T_2

print("F_2:",F_2)
```

```python
#Se generara un vector tiempo por rango y por numero de muestras:

  #Vector tiempo -> Cada numero de la lista sera un punto de el tiempo

#Discretizar tiempo

k = 200 #Número de muestras número de muestras que va mostrar
ti = 0 #Tiempo inicial
tf = 4*T_1 #Tiempo final -> El #4 significa los ciclos de la señal en la grafica, en otras palabras son las montañas.

#NOTA: si el ti empieza en -10 y tf = 10*T_1 -> Esto signica que habran 20 ciclos o montañas, ya que no empieza ti = 0 para que solo hayan 20 ciclos.

Fs = 20 * F_2 #swtich mas rapido que la velocidad de la señal
Ts = 1/Fs #Tiempo de muestreo
print(Ts)

#Crear vector discretizado

tv = np.arange(start=ti,stop=tf+Ts,step=Ts) #Discretizando el tiempo,         Ts = incremento entre cada número de la secuencia (Paso de tiempo).
print(tv)
print(tv.shape)

#Creando la señal, pero primero creando las dimensiones del arreglo

#Discretizar señal
xv = A_1*np.cos(W_1*tv) - (A_2*np.sin(W_2*tv+U))
print (xv)
print (xv.shape)
```

```python
import matplotlib.pyplot as plt

plt.plot(tv,xv)
plt.xlabel('tiempo [s]') #Nombre del eje x
plt.ylabel('$A\sin[\omega t_k]$') #Nombre del eje y
plt.show()
```

```python
#crear funcion en python
def modificar_amplitud(x,ymin = -5, ymax = 5): #en general se pueden definir valores por defecto.     Def -> Es para crear un nueva funcion
  '''
  Codigo base para simular proceso de cero y pendiente
  Se ingresa arreglo de numpy y los valores min  y max despues de cero y pendiente
  '''

#Hallando la pendiente

  xmax = max(x) # x.max() Esta funcion saca el maximo de la señal discretizada
  xmin = min(x) # x.min() Esta funcion saca el minimo de la señal discretizada

  m = (ymax - ymin)/(xmax - xmin) # Aqui se modifica la pendiente para escalar la señal, y que pase de 0 a 5 V por ejemplo

  c = ymin - m*xmin #Se halla c, porque la ecuacion y = mx+c lo requiere y para que la forma de la señal original se mantenga

  yv = m*x + c #Se calcula la señal discretizada de 0 a 5V
  return yv #Hasta aca va el codigo y imprime yv


```

```python
yv = modificar_amplitud(xv,ymin=0,ymax=10), #Donde xv es la señal que se quiere modificar, la cual se ajusta su rango de amplitud para que este entre ymin y ymax
print(yv) #Imprime
print(yv.shape) #Imprime la forma del arreglo unidimensional con el # de elementos que la señal original
```

```python
plt.subplot(121) #Donde 1 = # de filas, 2 = # de columnas, 1 = # Dibujo
plt.plot(tv,xv)
plt.xlabel('tiempo [s]') #Nombres de los ejes
plt.ylabel('$x[t_k]$')

plt.subplot(122)
plt.plot(tv,yv,'y') #Imprimir señal acondicionada
plt.xlabel('tiempo [s]')
plt.ylabel('$y[t_k]$')

plt.subplots_adjust(hspace=0.5,wspace=0.5) #ajustar espaciado entre gráficas
plt.show()
#A la izquierda se ve la señal original, a la derecha se ve la señal a 0 y 5V
```

```python
from scipy.spatial.distance import cdist
# crear vector tiempo por rango y por número de muestras
K = 200 # número de muestras
ti = 0 # tiempo incial referencia [s]
tf = 2*T_1 # tieimpo final  [s]
# vector de tiempo equiespacioado
fs = 20 * F_2
ts = 1/fs
tv = np.arange(start=ti,stop=tf+ts,step=ts)
y = A_1*np.cos(W_1*tv) - (A_2 * np.sin(W_2*tv+U))

ycs = ycs = modificar_amplitud(y, ymin=-3.3,ymax=3.3)#cero y pendient

#número de bits y vector de cuantización
nbits = 4
rmin = -3.3
rmax = 3.3
ve = np.linspace(rmin,rmax,2**nbits) #vector de cuantización

#distancias y cero pendiente a vector cuantizacion
d = cdist(ycs.reshape(-1,1),ve.reshape(-1,1))#se utiliza reshape(-1,1) para generar vectores columna como lo pide cdist
d.shape
print(d)
```

```python
#se requiere identificar el elemento ve[j] más cercano a y[i] para genera señal cuantizada
ind = np.argmin(d,axis=1) #el parámetro axis = 1 indica que busca la posición a lo largo de las columnas del elemento más pequeño en cada fila
ind
```

```python
print(ycs[1],ind[1],ve[ind[1]]) # para y[50] = 0.58, el estado más cercano en ve es el elemento 1 que corresponde a ve[1]=0.71
```

```python
#función cuantización punto a punto

def my_cuantizador(yn, vq) : #yn punto a #cuantizar, vq vector de estados

  Ne = vq.shape[0] #tamaño vector de estados
  dn = cdist(yn.reshape(-1,1),vq.reshape(-1,1))#distancia yn a vector estados, reshape(-1,1) asegura vectores columna para poder utilizar cdist
  ind = np.argmin(dn) #posición distancia min
  return vq[ind]
```

```python
#recorrer vector cero y spam discretizado para cuantizar

for i in range(ycs.shape[0]):
   if i==0: #primera muestra
     yq = my_cuantizador(ycs[i],ve)
   else:
     yq = np.append(yq,my_cuantizador(ycs[i],ve)) #apilar datos cuantizando si i>0
```

```python
#graficar señales

plt.plot(tv,ycs,c='r', label='ycs')#cs continua

plt.stem(tv,ycs, label='ycs_dis')#cs discreta

#estados de cuantización

for v in ve:
  plt.plot(tv,v*np.ones(len(tv)))

# graficar señal Digital (discreta + cuantizada)

plt.plot(tv,yq,c='y',marker='o',label='ydig', linewidth=2)

plt.legend()
plt.grid()
plt.xlabel('t')
plt.ylabel('Amplitud')
plt.show()
```

```python
def my_ceropen(x,ymin = -5, ymax = 5):

  xmax = max(x) # x.max() punto maximo de la entrada
  xmin = min(x) # x.min() punto maximo de la entrada
  m = (ymax - ymin)/(xmax - xmin)
  c = ymin - m*xmin # ecuacion de corte de la recta
  yv = m*x + c  # ecuacion de la recta
  return yv
```
